# **3. Metode Prediksi yang Digunakan**

Proyek ini menggunakan beberapa algoritma machine learning sebagai berikut:

1. **Random Forest Classifier**:  
   Algoritma berbasis ensemble yang membangun beberapa decision tree dari subset data yang berbeda dan menggabungkan hasilnya untuk meningkatkan akurasi dan mencegah overfitting. Cocok untuk menangani dataset dengan fitur yang kompleks.

2. **Decision Tree Classifier**:  
   Algoritma yang menggunakan struktur pohon untuk membuat keputusan berdasarkan aturan yang dihasilkan dari fitur dataset. Mudah dipahami dan diinterpretasikan, namun cenderung overfitting pada data besar.

3. **Support Vector Classifier (SVC)**:  
   Algoritma yang memanfaatkan hyperplane untuk memisahkan data ke dalam kelas yang berbeda dengan margin terbesar. Cocok untuk dataset dengan dimensi tinggi, meskipun kurang efisien pada dataset besar.

4. **K-Nearest Neighbors (KNN)**:  
   Algoritma berbasis instance-based learning yang mengklasifikasikan data berdasarkan kedekatannya dengan data tetangga terdekat. Sangat sederhana, namun performanya dipengaruhi oleh parameter k dan skala data.

5. **Logistic Regression**:  
   Algoritma linear untuk memprediksi probabilitas kelas target. Digunakan sebagai baseline karena memiliki performa baik pada dataset yang mudah dipisahkan secara linear.

6. **XGBoost Classifier**:  
   Algoritma ensemble berbasis boosting yang sangat kuat dan efisien untuk menangani dataset dengan jumlah data besar dan fitur kompleks. Memiliki kemampuan untuk mengurangi overfitting melalui regularisasi.

---

# **4. Metode Evaluasi**

Untuk mengevaluasi performa model, digunakan beberapa metrik evaluasi sebagai berikut:

1. **Akurasi**:  
   Mengukur persentase prediksi benar dari total prediksi.  
   Formula:  
   $$
   \text{Akurasi} = \frac{\text{Prediksi Benar}}{\text{Total Data}}
   $$

2. **Precision**:  
   Mengukur proporsi prediksi positif yang benar dari total prediksi positif. Berguna untuk menghindari *false positive*.  
   Formula:  
   $$
   \text{Precision} = \frac{\text{True Positive (TP)}}{\text{True Positive (TP)} + \text{False Positive (FP)}}
   $$

3. **Recall (Sensitivity)**:  
   Mengukur proporsi data positif yang teridentifikasi dengan benar oleh model. Berguna untuk menghindari *false negative*.  
   Formula:  
   $$
   \text{Recall} = \frac{\text{True Positive (TP)}}{\text{True Positive (TP)} + \text{False Negative (FN)}}
   $$

4. **F1-Score**:  
   Rata-rata harmonis dari precision dan recall. Berguna untuk mengevaluasi model pada data yang tidak seimbang.  
   Formula:  
   $$
   \text{F1-Score} = 2 \cdot \frac{\text{Precision} \cdot \text{Recall}}{\text{Precision} + \text{Recall}}
   $$

5. **Confusion Matrix**:  
   Matriks yang menunjukkan jumlah prediksi benar dan salah untuk setiap kelas. Mempermudah analisis kesalahan model dengan memberikan informasi tentang:
   - **True Positives (TP)**: Data positif yang diklasifikasi benar.  
   - **False Positives (FP)**: Data negatif yang salah diklasifikasi sebagai positif.  
   - **True Negatives (TN)**: Data negatif yang diklasifikasi benar.  
   - **False Negatives (FN)**: Data positif yang salah diklasifikasi sebagai negatif.

# **Modelling**

Membuat Fungsi Untuk Mendapatkan Hasil Evaluasi Masing-Masing Model

```python
def predict_and_plot(model, inputs, targets, name=''):
    # Melakukan prediksi
    preds = model.predict(inputs)

    # Menghitung metrik evaluasi
    accuracy = accuracy_score(targets, preds)
    precision = precision_score(targets, preds, average='weighted')
    recall = recall_score(targets, preds, average='weighted')
    f1 = f1_score(targets, preds, average='weighted')

    # Menampilkan metrik evaluasi
    print(f"{name} Metrics:")
    print("Accuracy: {:.2f}%".format(accuracy * 100))
    print("Precision: {:.2f}%".format(precision * 100))
    print("Recall: {:.2f}%".format(recall * 100))
    print("F1-Score: {:.2f}%".format(f1 * 100))
    print()

    # Membuat confusion matrix
    cf = confusion_matrix(targets, preds, normalize='true')
    plt.figure()
    sns.heatmap(cf, annot=True, cmap='Blues', fmt='.2f')
    plt.xlabel('Prediction')
    plt.ylabel('Target')
    plt.title('{} Confusion Matrix'.format(name))

    return preds
```

```python
train_preds = predict_and_plot(log_reg, X_train, y_train, 'Logistic Regression Train')

val_preds = predict_and_plot(log_reg, X_test, y_test, 'Logistic Regression Validation')
```
